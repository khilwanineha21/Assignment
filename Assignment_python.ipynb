```python
'''61) Write a Python function to calculate the factorial of a number (a 
nonnegative integer) '''
def factorial(number):
    factorial = 1
    if number < 1:
        print("invalid number")
    elif number == 0:
        print("the factorial of 0 is 1")
    else:
        for i in range(1,number+1):
            factorial = factorial * (i)
        return factorial

num = int(input("enter a number"))
print(f"factorial of {num} is {factorial(num)}")
    
```

```python
'''63)Write a Python function to check whether a number is perfect or not.'''
def perfect(number):
    sum = 0
    for i in range(1,number):
        if number % i == 0:
            sum = sum + i
    return sum == number
num = int(input("Enter a number"))
if perfect(num):
    print(f"{num} is a perfect number ")
else:
    print(f"{num} is not a perfect number")
```

```python
'''89) How Do You Handle Exceptions with Try/Except/Finally in Python? 
Explain with coding snippets.'''
try:
    a=int(input("enter the value"))
    b=int(input("enter the value"))
    c=a*b
except Exception as e:
    print(f"error occured {e}")
else:
    print(c)
finally:
    print("done")

''' finally:- always gets excecuted
try:- write the code which we want to excecute main code is written in try block
except:- if any error occurs in the try block it will be catched by the except block it will raise the error without doing any crash
else:- if their is no exception in try block'''
```
